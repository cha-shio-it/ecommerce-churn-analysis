```python
import pandas as pd
import numpy as np
# 读取数据
df = pd.read_csv(r'D:\My_First_Project\src\data\user_data.csv')

# 查看前五行
df.head()
```

```python
# 数据集的行列数
print(f"数据集形状：{df.shape}")

# 查看数据类型、缺失值
df.info()
```

```python
# 数值字段统计
df.describe()
```

```python
# 查看缺失值占比（%)
missing_pct = (df.isnull().sum() / len(df)) * 100
print(missing_pct)
```

```python
#处理缺失值
max_day = df['Days_Since_Last_purchase'].max()
df.fillna({
    'Session_Duration_Avg': 0,
    'Days_Since_Last_purchase': max_day+1,
    'Customer_Service_calls': 0,
    'Returns_Rate': 0,
    'Discount_Usage_Rate': 0,
    'Product_Reviews_Written': 0
}, inplace=True)
```

```python
# 处理重复行
duplicates = df.duplicated().sum()
print(f"重复行数: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
```

```python
# Total_Purchases总购买次数存在负数,将负数改为0
df.loc[df['Total_Purchases']<0,'Total_Purchases'] = 0

# Cart_Abandonment_Rate弃购率存在超过100，截断到[0,100]
df['Cart_Abandonment_Rate'] = df['Cart_Abandonment_Rate'].clip(0,100)

# Discount_Usage_Rate折扣使用率存在超过100,截断到[0,100]
df['Discount_Usage_Rate'] = df['Discount_Usage_Rate'].clip(0, 100)
```

```python
cols_to_convert = df.select_dtypes(include=[np.number]).columns
for col in cols_to_convert:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col], errors='coerce')
print(df.dtypes)
```

```python
# 整体流失率
churn_rate = df['Churned'].mean()
print(f"整体流失率: {churn_rate:.2%}")
```

```python
df.to_csv(r'D:\My_First_Project\src\data\cleaned_user_data.csv', index=False)
print("清洗后的数据已保存为 cleaned_user_data.csv")
```
